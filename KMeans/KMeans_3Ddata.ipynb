```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
centroids = [(-5,-5,5),(5,5,-5),(-2.5,2.5,4),(2.5,-2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=300,cluster_std=cluster_std,centers=centroids,n_features=4,random_state=3)
X
```

```python
import plotly.express as px
```

```python
fig=px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
fig.show()
```

```python
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit_predict(X)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=4)
y_pred=km.fit_predict(X)
```

```python
df=pd.DataFrame()
df['col1']=X[:,0]
df['col2']=X[:,1]
df['col3']=X[:,2]
df['lebel']=y_pred
```

```python
fig=px.scatter_3d(df,x='col1',y='col2',z='col3',color='lebel')
```

```python
fig.show()
```
